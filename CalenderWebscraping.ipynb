```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
data = requests.get('https://www.ccny.cuny.edu/registrar/fall')
```

```python
bSoup = BeautifulSoup(data.text, 'html.parser')
print(bSoup.title)
```

```python
#bSoup.prettify
```

```python
elements = bSoup.find_all('table')
print(elements[:100])
```

```python
all_table = bSoup.find_all('table')
print(len(all_table))
```

```python
calender = [table for table in all_table]
len(calender)
```

```python
from IPython.display import HTML
HTML(str(calender[0]))
```

```python
rows = [ row for row in calender[0].find_all('tr')]
col_heads =[col.get_text() for col in  rows[0].find_all('th')]
```

```python
value_rows = [[col.get_text().strip() for col in row.find_all('td')] for row in  rows[1:]]
series_list = list(zip(*value_rows))
```

### Renaming the columns and using "Date" as the index

```python
rename_col_heads = ['DATE', 'DOW', 'TEXT WITH EXPLANATION']
df = pd.DataFrame(dict(zip(rename_col_heads, series_list)))                                                                    
```

```python
df.set_index('DATE', inplace = True)    
```
